### Bloco A - Operadores e variáveis
- Dados ano de nascimento e morte de 4 autores (Machado (1839-1908), Eça (1845-1900), Guimarães Rosa (1908-1967), Clarice (1920-1977)), calcule a idade de cada um em f-string.
- Machado publicou um livro em 1881 e outro em 1899.
Entre o primeiro e o último livro, há 2 intervalos.
Calcule em variáveis:

    1. Quantos anos se passaram de 1881 até 1899.

    2. Quanto vale cada intervalo, se esse tempo for dividido em 2 partes.

- Se um romance tem 400 páginas e o aluno lê 15 por dia, quantos dias inteiros precisa? E quantas páginas sobram no último dia?

```python
# Anos de nascimento e morte
machado_idade = 1908 - 1839
eca_idade = 1900 - 1845
rosa_idade = 1967 - 1908
clarice_idade = 1977 - 1920

print(f"Idade de Machado: {machado_idade} anos")
print(f"Idade de Eça: {eca_idade} anos")
print(f"Idade de Guimarães Rosa: {rosa_idade} anos")
print(f"Idade de Clarice: {clarice_idade} anos")
```

```python
# Intervalo entre publicações
ano_inicio = 1881
ano_fim = 1899

anos_passados = ano_fim - ano_inicio
tamanho_intervalo = anos_passados / 2

print(f"1. Se passaram {anos_passados} anos de 1881 até 1899.")
print(f"2. Cada intervalo vale {tamanho_intervalo} anos.")
```

```python
# Páginas por dia + sobra no último dia
total_paginas = 400
paginas_por_dia = 15

dias_inteiros = total_paginas // paginas_por_dia
paginas_sobra = total_paginas % paginas_por_dia

print(f"O aluno precisa de {dias_inteiros} dias inteiros.")
print(f"E sobram {paginas_sobra} páginas no último dia.")
```

### Bloco C - Strings
- Dada a frase `"Words are, in my not-so-humble opinion, our most inexhaustible source of magic"`, aplique e guarde em diferentes variáveis: letras minusculas, letras maiusculas e substitua "Words" por "Spells".
- Construa uma f-string descrevendo um autor com três variáveis: nome, ano de nascimento, obra principal.
- Dada a palavra `"onomatopeico"`, verifique com `in` se ela contém as letras `"o"`, `"z"` e `"a"`. Três checks.


```python
# Words are, in my not-so-humble (...)
frase = "Words are, in my not-so-humble opinion, our most inexhaustible source of magic"

frase_minuscula = frase.lower()
frase_maiuscula = frase.upper()
frase_substituida = frase.replace("Words", "Spells")

print("Minúsculas:", frase_minuscula)
print("Maiúsculas:", frase_maiuscula)
print("Substituída:", frase_substituida)
```

```python
# f-string: nome, ano, obra
nome = "J.R.R. Tolkien"
ano_nascimento = 1892
obra = "O Senhor dos Anéis"

descricao = f"O autor {nome} nasceu em {ano_nascimento} e é conhecido por escrever {obra}."
print(descricao)
```

```python
# onomatopeico
palavra = "onomatopeico"

tem_o = "o" in palavra
tem_z = "z" in palavra
tem_a = "a" in palavra

print(f"Contém 'o': {tem_o}")
print(f"Contém 'z': {tem_z}")
print(f"Contém 'a': {tem_a}")
```
