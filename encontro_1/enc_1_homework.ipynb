## Exercícios por bloco

### Bloco A - Operadores e variáveis
- Dados ano de nascimento e morte de 4 autores (Machado, Eça, Guimarães Rosa, Clarice), calcule a idade de cada um em f-string.
- Machado publicou Memórias Póstumas em 1881, Quincas Borba em 1891, Dom Casmurro em 1899. Qual o intervalo médio entre publicações?
- Se um romance tem 400 páginas e o aluno lê 15 por dia, quantos dias inteiros precisa? E quantas páginas sobram no último dia? 

### Bloco B - Tipos e conversões
- Dada a string `"1899"`, converta para `int` e some 10 anos.
- Aplique `type()` a 5 valores diferentes (incluindo `None`, um float, um bool).
- **Pergunta conceitual**: por que `"10" + "5"` não dá 15? Explique em duas frases.

### Bloco C - Strings
- Dada a frase `"Dom Casmurro é um romance de Machado"`, aplique `.lower()`, `.upper()`, `.replace("Machado", "Assis")`. Três linhas, três transformações.
- Construa uma f-string descrevendo um autor com três variáveis: nome, ano de nascimento, obra principal.
- Dada a palavra `"onomatopeico"`, verifique com `in` se ela contém as letras `"o"`, `"z"` e `"a"`. Três checks.
- Dada `titulo = "Memórias Póstumas"`, imprima a primeira letra, a última, e as letras da posição 9 à 15 (slicing).

### Bloco D - Listas e slicing
- Dada a lista `tokens = ["o", "cão", "correu", "no", "parque", "grande"]`, faça: primeiros 3, últimos 2, invertida, a cada 2.
- Dadas duas listas (substantivos e verbos), concatene com `+` e conte o total com `len()`.
- **Pergunta conceitual**: qual a diferença entre `tokens[0]` e `tokens[-1]`? Por que a contagem começa em zero?

### Bloco E - Dicionários
- Crie um dicionário com 5 palavras e suas frequências. Acesse 2 valores diretamente. Use `.get()` para uma chave inexistente.
- Adicione uma palavra nova. Sobrescreva a frequência de uma existente.
- Verifique com `in` se `"Machado"` está nas chaves.
- Imprima `.keys()`, `.values()`, `.items()` separadamente.

### Bloco F - Tuplas
- Crie uma tupla `(autor, ano, obra)`. Faça unpacking em três variáveis com uma linha só.
- Tente alterar um elemento da tupla. Copie a mensagem de erro que Python devolve e explique em uma frase o porquê.

## Desafio opcional - extra challenge

> **Análise da sua frase favorita.** Escolha uma frase qualquer (não precisa ser Machado). 
> 1. Guarde a frase numa variável e converta para minúsculas com `.lower()`.
> 2. Separe em tokens com `.split()` e guarde numa lista.
> 3. Conte quantos tokens tem usando `len()`.
> 4. Crie um dicionário onde cada palavra é chave e o valor é o número de letras dela (para uma frase de 6-8 palavras, dá para fazer manualmente).
> 5. Olhando o dicionário, identifique a palavra mais longa por inspeção.
> 6. Escreva um resumo em f-string: *"A frase de X tem N tokens; a mais longa é 'Y' com Z letras."*
