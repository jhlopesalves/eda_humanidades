```python
def top_n(frequencias: dict, n: int = 10) -> list:
    """
    Retorna os n itens mais frequentes de um dicionário de frequências.

    Parâmetros
    ----------
    frequencias : dict
        Dicionário onde as chaves são tokens (str) e os valores
        são suas contagens (int).
    n : int, opcional
        Quantidade de itens a retornar. Padrão: 10.

    Retorna
    -------
    list
        Lista de tuplas (token, contagem), ordenada da maior
        para a menor contagem.

    Exemplo
    -------
    >>> freq = {"casa": 5, "rua": 2, "céu": 8}
    >>> top_n(freq, n=2)
    [('céu', 8), ('casa', 5)]
    """
    return sorted(frequencias.items(), key=lambda par: par[1], reverse=True)[:n]
```
