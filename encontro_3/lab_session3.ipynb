### 1. Tokens e tipos: a distinção fundamental!

Antes de qualquer métrica, dois conceitos:

- **Token** == cada *ocorrência* de palavra no texto. Se "autor" aparece 50 vezes, são 50 tokens.
- **Tipo** (ou *palavra única*, ou *vocabulário*) == cada palavra *distinta*. "Casa" aparecendo 50 vezes conta como **1 tipo**.


> Toda métrica que vem a seguir é uma razão entre tokens e tipos, ou uma forma de olhar para a distribuição deles.

## 2. TTR — Type-Token Ratio (Razão Tipo/Token)

**O que mede:** *diversidade lexical*, o quão variado é o vocabulário em relação ao tamanho do texto.

**Fórmula:**

$$
\text{TTR} = \frac{\text{tipos}}{\text{tokens}} = \frac{V}{N}
$$

**Um exemplo:**

$$
\text{TTR} = \frac{9.525}{51.408} \approx 0{,}1853
$$

```python
print(f"TTR: {metricas['ttr']:.4f}")
```

**Leitura:** a cada 100 palavras corridas, cerca de **18,5 são "novas"** em média (estatisticamente; não significa que de fato a cada 100 aparece esse número exato).

**Cuidado importante:**

> TTR **depende fortemente do tamanho do texto**. Textos longos têm TTR mecanicamente menor; não porque o autor seja "pobre" de vocabulário, mas porque o vocabulário cresce de forma **sublinear** em relação ao número de palavras (fenômeno conhecido como **Lei de Heaps** -> falaremos mais sobre isso mais tarde no curso).

Por isso, **não dá pra comparar diretamente** o TTR de um romance com o de um conto. Para comparações mais justas, existem variantes padronizadas (MATTR, MTLD, vocd-D) que controlam pelo tamanho.

Faixas típicas (São apenas interpretações comuns, podem variar bastante a depender do objeto de análise):

| Tamanho do texto | TTR esperado |
|---|---|
| ~1.000 tokens (texto curto) | 0,40 – 0,50 |
| ~5.000 tokens (conto) | 0,25 – 0,35 |
| ~50.000 tokens (romance) | 0,15 – 0,22 |

## 3. Hapax legomena

**Etimologia:** do grego ἅπαξ λεγόμενον - "*dito uma só vez*". O termo vem da filologia clássica, usado por estudiosos de Homero para marcar palavras que aparecem uma única vez no corpus homérico.

**Definição:** palavra cuja frequência no texto é exatamente **1**.

**Nosso exemplo:**

```python
print(f"Hapax legomena: {metricas['hapax']:,}")
print(f"Razão hapax: {metricas['razao_hapax']:.4f}")
```

$$
\frac{\text{hapax}}{\text{tipos}} = \frac{5.466}{9.525} \approx 0{,}5739
$$

**Leitura:** cerca de **57% do vocabulário** do livro é formado por palavras que aparecem **uma única vez**.

**Isso é muito?** É **exatamente o esperado**. A distribuição de frequências de palavras em qualquer língua natural costuma ter uma "cauda longa" - a grande maioria das palavras do vocabulário é rara. Razões hapax entre 40% e 60% são típicas em literatura. Esse padrão é comum em **qualquer corpus suficientemente grande, em qualquer língua, de qualquer autor** (Fan, 2010).

**Por que isso é interessante?**

- **Linguisticamente:** muitos hapax são nomes próprios, neologismos, termos técnicos ou palavras usadas num contexto muito específico. Eles carregam alta carga informacional.
- **Computacionalmente:** a abundância de hapax é justamente o motivo pelo qual modelos de linguagem precisam de técnicas de *smoothing* — eles têm que lidar com palavras vistas pouquíssimas vezes (ou nunca vistas).
- **Filologicamente:** hapax servem como pistas para datação de textos, atribuição de autoria e identificação de interpolações.
- **Metodologicamente:** em corpora grandes, hapax também são bons objetos de inspeção, porque podem revelar erros de processamento, problemas de OCR, typos, tokenização inadequada ou palavras mal normalizadas. Nem todo hapax é um achado linguístico: alguns são sinais de que o pipeline de limpeza precisa ser revisado.

Duas observações importantes:

### Observação linguística: são todas palavras funcionais

Em terminologia técnica, são **palavras de classe fechada** (*closed-class words*) ou ***function words***:

- Conjunções/complementizadores: *que*
- Preposições e contrações: *de, com, do, da*
- Artigos: *um, uma, os*
- Advérbio de negação: *não*
- Verbo de ligação no imperfeito: *era*

Elas são o **esqueleto gramatical** da língua - não dizem nada sobre o *tema* do livro. É por isso que, em análise de tópicos, costuma-se filtrá-las com listas de ***stopwords***. Mas atenção: para análise de **estilo** ou **atribuição de autoria**, essas palavras são, na verdade, **as mais úteis**; porque revelam padrões inconscientes do autor.

### Observação estatística: a Lei de Zipf

A **Lei de Zipf** prevê que, ordenando as palavras por frequência (rank 1 = mais frequente), a frequência decai aproximadamente como:

$$
f(r) \approx \frac{C}{r^{s}}, \quad s \approx 1
$$

Em linguagem comum: a 2ª palavra mais frequente aparece cerca de **metade** das vezes que a 1ª; a 3ª, cerca de **um terço**; e assim por diante.

Veja isso no nosso livro (previsão com \(s=1\), tomando \(f_1 = 2.177\)):

| rank | palavra | observado | previsto por Zipf |
|---|---|---|---|
| 1 | que | 2.177 | 2.177 |
| 2 | de | 2.128 | 1.089 |
| 3 | não | 1.112 | 726 |
| 4 | um | 1.023 | 544 |
| 5 | do | 743 | 435 |
| 10 | os | 492 | 218 |

O **decaimento existe** (qualitativamente Zipf bate), mas o **topo desvia**; "de" deveria ser metade de "que", e está praticamente igual. Esse desvio no topo é **bem documentado**: Zipf é uma aproximação melhor no meio da distribuição; nos extremos há desvios sistemáticos. Existem modelos refinados (Zipf-Mandelbrot) que corrigem isso.

### Observação estilística (machadiana)

- **"que" liderando "de"** sugere alta densidade de subordinação (orações relativas, completivas) - coerente com o estilo do Machado, conhecido pelos períodos longos e encaixes parentéticos.
- **"era" no top 10** é uma assinatura narrativa: imperfeito do indicativo. Machado *descreve* cenas e estados passados. Se "foi" liderasse no lugar de "era", o pulso narrativo do livro seria outro (mais perfectivo).
