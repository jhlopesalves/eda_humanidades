````python
mertonian_norms = """
The norms are expressed in the form of prescriptions, proscriptions, preferences, and permissions. They are legitimatized in terms of institutional values. These imperatives, transmitted by precept and example and reenforced by sanctions are in varying degrees internalized by the scientist, thus fashioning his scientific conscience or, if one prefers the latter-day phrase, his super-ego… [This scientific ethos] can be inferred from the moral consensus of scientists as expressed in use and wont, in countless writings on the scientific spirit and in moral indignation directed toward contraventions of the ethos.

An examination of the ethos of modern science is only a limited introduction to a larger problem: the comparative study of the institutional structure of science. Although detailed monographs assembling the needed comparative materials are few and scattered, they provide some basis for the provisional assumption that "science is afforded opportunity for development in a democratic order which is integrated with the ethos of science." This is not to say that the pursuit of science is confined to democracies.
"""

texto_limpo = (
    mertonian_norms.lower()
    .replace(",", " ")
    .replace(".", " ")
    .replace("–", " ")
    .replace("\n", " ")
    .replace("[", " ")
    .replace("]", " ")
)

tokens_mertonian = texto_limpo.split()

print(tokens_mertonian[:100])
print(len(tokens_mertonian))
````

### A resposta mais votada foi a seguinte:

```python
# From an efficiency perspective, you're not going to beat:
s.translate(None, string.punctuation)

# For higher versions of Python use the following code:
s.translate(str.maketrans('', '', string.punctuation))

# It's performing raw string operations in C with a lookup table
# there's not much that will beat that but writing your own C code.

# However,
# If speed isn't a worry, another option though is:
```

This is faster than `s.replace` with each char, but won't perform as well as non-pure python approaches such as regexes or `string.translate`, as you can see from the below timings. For this type of problem, doing it at as low a level as possible pays off.

## 3. Retomando dicionários

````python
# Explicito, mais transparente e um pouco mais fácil de ler para iniciantes
frequencias = {}

# Itera sobre cada token (palavra) na lista tokens_mertonian
for token in tokens_mertonian:
    # Verifica se o token já é uma chave existente no dicionário 'frequencias'
    if token in frequencias:
        # Se existir, acessa o valor atual e incrementa (soma) 1 à contagem
        frequencias[token] += 1
    # Caso o token ainda não exista no dicionário
    else:
        # Cria a nova chave com o token e inicializa sua contagem com 1
        frequencias[token] = 1

print(frequencias)
````

### `.get()`

> "Pegue a frequência atual do token. Se ele ainda não existir, considere que a frequência atual é 0. Depois some 1."

O `.get(chave, valor_padrão)` evita o `if/else` e é considerado mais idiomático.

````python
# Mais idiomático, conciso, econômico; considerado mais maduro
frequencias = {}

# Para cada token na lista tokens_mertonian
for token in tokens_mertonian:
    # Define ou atualiza a chave 'token' em 'frequencias'
    # .get(token, 0) retorna o valor atual se 'token' existir, senão retorna 0
    # Em seguida, soma 1 a esse valor e atribui à chave 'token'
    frequencias[token] = frequencias.get(token, 0) + 1

display(list(frequencias.items())[:10])
````

### `.items()`, unpacking e mais controle sobre o dicionário

`.items()` devolve pares `(chave, valor)`. Com unpacking, podemos iterar sobre os dois ao mesmo tempo.

````python
# Iterar sobre chave e valor ao mesmo tempo: .items()
for palavra, contagem in frequencias.items():
    if len(palavra) > 8 and contagem > 1:
        print(f"{palavra}: {contagem}")
````

### 4.1 TTR - Type-Token Ratio

A **TTR** é uma métrica clássica de **diversidade lexical**:

$$
\text{TTR} = \frac{\text{número de tipos distintos}}{\text{número total de tokens}}
$$

Quanto mais próximo de 1, mais variado o vocabulário em relação ao tamanho do texto. Atenção: TTR depende do **tamanho do texto**. Comparações justas exigem textos de tamanhos similares.

````python
valor_ttr = len(set(tokens_mertonian)) / len(tokens_mertonian)
print(f"TTR: {valor_ttr:.3f}")
````

### 4.2 Amplitude (range)

A amplitude é a diferença entre o maior e o menor valor observado:

$$
\text{amplitude} = \max(x) - \min(x)
$$

É a medida de dispersão mais simples possível. Sua fraqueza: depende **apenas dos extremos**. Um único valor atípico distorce a amplitude inteira.

````python
# Construindo a lista de comprimentos com loop explícito
comprimentos = []

for token in tokens_mertonian:
    comprimentos.append(len(token))

print(comprimentos[:20])
print(f"Total de comprimentos: {len(comprimentos)}")
````

````python
amplitude = max(comprimentos) - min(comprimentos)
print(f"Menor palavra: {min(comprimentos)} caracteres")
print(f"Maior palavra: {max(comprimentos)} caracteres")
print(f"Amplitude: {amplitude}")
````

### 4.3 Mediana

A **mediana** é o valor central de um conjunto ordenado. Para encontrá-la, precisamos:

1. Ordenar os valores.
2. Se $(n)$ é ímpar, pegar o valor da posição central.
3. Se $(n)$ é par, tirar a média dos dois valores centrais.

A mediana é o exemplo mais simples de **estatística resistente** - vocabulário introduzido por Tukey para descrever medidas pouco afetadas por valores extremos. Se uma palavra de 30 caracteres aparecesse no nosso texto, a média subiria visivelmente; a mediana, quase nada.

A escolha entre média e mediana depende do que queremos representar e de quão sensível a distribuição é a outliers.

````python
# Caso par: há dois valores centrais
valores = [3, 7, 2, 9]

valores_ordenados = sorted(valores)
indice_do_meio = len(valores_ordenados) // 2

mediana = (
    valores_ordenados[indice_do_meio - 1]
    + valores_ordenados[indice_do_meio]
) / 2

print(valores_ordenados)
print(mediana)
````

````python
# Caso ímpar: há um único valor central
valores = [3, 7, 2, 9, 5]

valores_ordenados = sorted(valores)
indice_do_meio = len(valores_ordenados) // 2

mediana = valores_ordenados[indice_do_meio]

print(valores_ordenados)
print(mediana)
````

## 5. Funções & blocos reutilizáveis

Uma função simples tem três partes:

1. nome;
2. entrada;
3. resultado devolvido com `return`.

A **média aritmética** é a soma de todos os valores dividida pela quantidade total de itens observados:

$$
\bar{x} = \frac{\sum_{i=1}^{n} x_i}{n} = \frac{x_1 + x_2 + \dots + x_n}{n}
$$

````python
def calcular_media(valores: list[float]) -> float:
    resultado = sum(valores) / len(valores)
    return resultado
````

### Fórmula da mediana:

A **mediana** é o valor central que divide ao meio um conjunto de dados ordenados ($x_1, x_2, \dots, x_n$). Sua fórmula depende se o número total de elementos ($n$) é par ou ímpar:

$$
\text{mediana} = \begin{cases}
  x_{\frac{n+1}{2}} & \text{se } n \text{ é ímpar} \\
  \frac{x_{\frac{n}{2}} + x_{\frac{n}{2} + 1}}{2} & \text{se } n \text{ é par}
\end{cases}
$$

````python
def calcular_mediana(valores: list[float]) -> float:
    valores_ordenados = sorted(valores)
    n = len(valores_ordenados)
    indice_do_meio = n // 2

    if n % 2 != 0: # Caso ímpar
        return valores_ordenados[indice_do_meio]
    else: # Caso par
        meio_esquerda = valores_ordenados[indice_do_meio - 1]
        meio_direita = valores_ordenados[indice_do_meio]
        return (meio_esquerda + meio_direita) / 2
````

$$
\text{TTR} = \frac{\text{número de tipos distintos}}{\text{número total de tokens}}
$$


````python
def calcular_ttr(tokens: list[str]) -> float:
    resultado = len(set(tokens)) / len(tokens)
    return resultado
````
