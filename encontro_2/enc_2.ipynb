```python
# Importamos a biblioteca 'requests', especializada em buscar informações na internet
import requests

# Definimos a URL do livro (formato .txt puro)
url_do_livro = "https://www.gutenberg.org/files/55752/55752-0.txt"

# Fazemos o download e extraímos apenas o texto
resposta = requests.get(url_do_livro)
livro_completo = resposta.text

print("Download concluído!")
```

## 1. O Texto como Sequência: Indexing e Slicing

Agora temos um livro inteiro guardado na variável `livro_completo`. Python enxerga esse texto gigantesco (String) como uma longa fita contendo milhares de caracteres em sequência.

Na análise de dados, frequentemente precisamos de uma "amostra". Não queremos olhar para o todo, mas sim extrair um parágrafo, as primeiras 100 palavras, ou o último capítulo. 

Para extrair fatias dessa sequência, usamos o conceito de **Slicing** (Fatiamento), operado através de colchetes `[]`.

**A regra do Slicing:**
* O computador sempre começa a contar do zero. O primeiro caractere está na posição `0`.
* A sintaxe é: `variavel[inicio:fim]`.
* O Python inclui o número inicial, mas **exclui** o número final. Se você pedir `[0:10]`, ele trará da posição 0 até a 9.

```python
# Fatiando a string: extraindo os primeiros 500 caracteres
# EXPERIMENTO: Altere o número 500 para 1000, 2000, 5000 e rode a célula novamente.
trecho_inicial = livro_completo[0:100]

print(trecho_inicial)
```

## 2. De Letras para Palavras: O Vocabulário do Autor

Até agora, nosso `livro_completo` é visto pelo Python como uma fita contínua de caracteres. Ele não sabe o que é uma palavra; apenas enxerga letras e espaços enfileirados.

Para analisar o texto, precisamos quebrar essa fita em unidades significativas: as palavras. Fazemos isso com o método `.split()`, que corta o texto em cada espaço em branco e guarda os pedaços em uma **Lista** (`list`).

### Types vs. Tokens (Listas vs. Sets)

Uma **Lista** guarda tudo exatamente na ordem em que aparece, incluindo repetições. Se o artigo "o" aparece cinco mil vezes no livro, a lista guardará cinco mil itens "o". O tamanho dessa lista nos dá o total de palavras do livro (na linguística, chamamos isso de *Tokens*).

Mas e se quisermos descobrir a riqueza do vocabulário do autor? Quantas palavras **únicas** ele utilizou para escrever a obra inteira?

Aqui introduzimos uma nova estrutura de dados: o **Set** (Conjunto). 
Assim como na matemática, um `set` em Python tem uma regra de ouro: **ele não aceita itens duplicados e não possui uma ordem fixa**. Se você converter a nossa lista gigante em um set, o Python varre o texto e descarta todas as repetições instantaneamente. O resultado é o nosso vocabulário único (os *Types*).

```python
# 1. Cortando a string gigante em uma Lista de palavras (Tokens)
# O método split() é uma ferramenta embutida nas strings
lista_de_palavras = livro_completo.split()

# 2. Convertendo a Lista em um Set para extrair palavras únicas (Types)
# set() é uma função solta do Python, assim como type() ou len()
vocabulario_unico = set(lista_de_palavras)

# 3. Comprovando a diferença estrutural
print("Total de palavras no texto (Tokens):", len(lista_de_palavras))
print("Tamanho do vocabulário único (Types):", len(vocabulario_unico))
```

## 3. Controle de Fluxo

Na análise de um texto, nem todas as palavras têm o mesmo peso analítico. Palavras como "o", "a", "de" e "para" (chamadas de *stop words*) são essenciais para a gramática, mas frequentemente atrapalham a análise de frequência e os cálculos de vocabulário. 

Precisamos de uma forma de dizer ao Python: *"Avalie esta palavra. Se ela for importante, guarde. Se for uma preposição, descarte."*

Fazemos isso através de estruturas condicionais, que funcionam como os "porteiros" do nosso código.

* **`if` (Se):** A primeira verificação. "Se a condição for verdadeira, faça isso".
* **`elif` (Ou se):** Uma condição alternativa, caso a primeira seja falsa.
* **`else` (Senão):** O último recurso. "Se nenhuma das anteriores for verdadeira, faça isso por padrão".
