## 5. Dicionários

Uma estrutura de dados que guarda valores no formato: `chave: valor`

As `chaves` precisam ser imutáveis, os `valores` podem ser heterogêneos.

```python
frequencias = {
    "verbo": 25,
    "substantivo": 10,
    "artigo": 29,
}

print(frequencias)
```

```python
print("verbo" in frequencias)
print("substantivo" in frequencias)
```

```python
lista_de_tags = ["verbo", "substantivo", "preposição", "artigo"]

frequencias = {
    "verbo": 25,
    "substantivo": 10,
    "artigo": 29,
}

for tag in lista_de_tags:
    print(f"tag: {tag}: {frequencias[tag]}")
```

```python
lista_de_tags = ["verbo", "substantivo", "preposição", "artigo"]

frequencias = {
    "verbo": 25,
    "substantivo": 10,
    "artigo": 29,
}

for tag in lista_de_tags:
    print(f"tag: {tag}: {frequencias.get(tag, "Valor não encontrado")}")
```
