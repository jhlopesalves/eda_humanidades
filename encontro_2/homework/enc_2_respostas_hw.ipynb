A média aritmética é calculada somando todos os valores de um conjunto de dados e dividindo o resultado pelo número total de elementos.

$$ \bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i = \frac{x_1 + x_2 + \dots + x_n}{n} $$

Calcule a média dos valores da lista e imprima o resultado usando uma f-string. Inclua um texto junto com o resultado.

```python
valores = [1, 2.2, 5, 10, 12, 15, 20, 5, 32, 33, 37, 42.42222222]

# Calcule N, ou seja, a quantidade de itens da lista.
quantidade_itens = len(valores)

# Calcule a soma de todos os itens.
soma_itens = sum(valores)

# Divida a soma dos itens (sigma) pelo número de itens (N)... pssst: use divisão comum... pssssssssst: com casas decimais e tudo mais =)
media = soma_itens / quantidade_itens

# Imprima uma f-string comunicando o resultado de forma clara (texto + o resultado)
print(f"A média aritmética da lista 'valores' é: {media}")
```

OPÇÃO 1 - Chaining dentro de parenteses

```python
cudos_ethical_science = """
Communalism (le communalisme): Scientific knowledge should be shared openly and regarded as a common heritage.

Universalism (l’universalisme): Truth claims should be evaluated independently of a scientist’s race, gender, status, or other personal characteristics.

Disinterestedness (le désintéressement): Scientists should act for the benefit of the collective scientific enterprise rather than for personal gain.

Organized Skepticism (le scepticisme organisé): Scientific findings should be subjected to rigorous scrutiny and peer review before being accepted.
"""

# 1. Substitua (),.’: e . por strings vazias ("") numa só linha utilizando parenteses
cudos_sem_pontuacao = (
    cudos_ethical_science.replace("(", "")
    .replace(")", "")
    .replace(",", "")
    .replace("’", "")
    .replace(":", "")
    .replace(".", "")
)

# 2. Normalize todo o texto para minúsculas.
cudos_minusculo = cudos_sem_pontuacao.lower()

# 3. Converta a string em uma lista.
lista_tokens_cudos = cudos_minusculo.split()

tokens_cudos_maior_que_um = []
# 4. Percorra os itens da lista usando um loop for.
for token in lista_tokens_cudos:
    # 5. Verifique se cada string tem mais de 1 caracter.
    if len(token) > 1:
        # 6. Se tiver, salve a string em uma lista; caso contrário, ignore-a.
        tokens_cudos_maior_que_um.append(token)
# 7. Imprima os 10 primeiros itens da lista.
print(tokens_cudos_maior_que_um[:10])
```

OPÇÃO 2 - Um 'replace' por vez

```python
cudos_ethical_science = """
Communalism (le communalisme): Scientific knowledge should be shared openly and regarded as a common heritage.

Universalism (l’universalisme): Truth claims should be evaluated independently of a scientist’s race, gender, status, or other personal characteristics.

Disinterestedness (le désintéressement): Scientists should act for the benefit of the collective scientific enterprise rather than for personal gain.

Organized Skepticism (le scepticisme organisé): Scientific findings should be subjected to rigorous scrutiny and peer review before being accepted.
"""

# 1. Substitua (),.’: e . por strings vazias ("")-> um replace por vez
cudos_sem_pontuacao = cudos_ethical_science.replace("(", "")
cudos_sem_pontuacao = cudos_sem_pontuacao.replace(")", "")
cudos_sem_pontuacao = cudos_sem_pontuacao.replace(",", "")
cudos_sem_pontuacao = cudos_sem_pontuacao.replace("’", "")
cudos_sem_pontuacao = cudos_sem_pontuacao.replace(":", "")
cudos_sem_pontuacao = cudos_sem_pontuacao.replace(".", "")


# 2. Normalize todo o texto para minúsculas.
cudos_minusculo = cudos_sem_pontuacao.lower()

# 3. Converta a string em uma lista.
lista_tokens_cudos = cudos_minusculo.split()

tokens_cudos_maior_que_um = []
# 4. Percorra os itens da lista usando um loop for.
for token in lista_tokens_cudos:
    # 5. Verifique se cada string tem mais de 1 caracter.
    if len(token) > 1:
        # 6. Se tiver, salve a string em uma lista; caso contrário, ignore-a.
        tokens_cudos_maior_que_um.append(token)
# 7. Imprima os 10 primeiros itens da lista.
print(tokens_cudos_maior_que_um[:10])
```

### Bloco C — Frequências com dicionário
A partir da lista de tokens filtrada no Bloco B:
1. Crie um dicionário vazio frequencias.
2. Percorra a lista de tokens com um `for`.
3. Para cada token, atualize o dicionário usando `.get(token, 0) + 1.`
4. Imprima o dicionário completo.
5. Imprima quantos types distintos existem (ou seja, `len(frequencias)`).
6. Imprima a frequência da palavra `scientific` e da palavra `science`.
7. Comente em uma linha o que esse contraste sugere (inclua uma célula markdown com sua resposta).

```python
# 1. Crie um dicionário vazio frequencias
frequencias = {}

# 2. Percorra a lista de tokens com um for.
for token in tokens_cudos_maior_que_um:
    # 3. Para cada token, atualize o dicionário usando .get(token, 0) + 1.
    frequencias[token] = frequencias.get(token, 0) + 1

# 4. Imprima o dicionário completo.
print(f"Dicionário completo:\n{frequencias}")

# 5. Imprima quantos types distintos existem (ou seja, len(frequencias)).
print(f"Quantidade de types distintos: {len(frequencias)}")

# 6. Imprima a frequência da palavra scientific e da palavra science.
print(f"Frequência da palavra 'scientific': {frequencias.get('scientific', 0)}")
print(f"Frequência da palavra 'science': {frequencias.get('science', 0)}")
```

Resposta conceitual:

1. **Análise Linguística (O texto):** A ausência da palavra `science` e a presença de `scientific` (3 vezes) sugerem que o texto foca mais na aplicação, características ou produtos da ciência (como 'scientific knowledge' e 'scientific findings') do que na ciência como um conceito abstrato ou isolado.

2. **Reflexão Metodológica (O código):** Embora a ausência da palavra seja um dado linguístico importante, a *ocorrência 0* exibida no resultado é completamente sintética. Foi um argumento arbitrário (`.get('science', 0)`) que demos ao Python para evitar um erro. Em contextos linguísticos, devemos ter cautela: o valor `0` é um retorno que nós injetamos no código para lidar com a ausência do token, e não uma contagem real de algo que existe no corpus.
